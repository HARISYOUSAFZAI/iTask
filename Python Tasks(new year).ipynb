```python
#15 . write a func bywhcih i can send a mail to anyone 
import email #emails formats
import smtplib #simple message transfer protocol library
import bs4
from email.mime.multipart import MIMEMultipart as mm
from email.mime.text import MIMEText as txt
from email.mime.base import MIMEBase as bse


def sendemail(emailid,pswd,To,sub,mesg,attachement):
    '''
    sendemail(emailid,pswd,sub,mesg,attachement) sends emails 
    emailid: (str) (work on gmail,yahoo and outlook accounts)
    pswd:(str)
    To: (str) (sender email id)
    mesg: (str) (msg want to deliver)
    attachement: (list) (files to be delivered)
    
    '''
    idlist=['gmail.com','yahoo.com','outlook.com']
    if emailid.partition("@")[2] == idlist[0]:
        server= smtplib.SMTP('smtp.gmail.com',587)
    elif emailid.partition("@")[2] == idlist[1]:
        server= smtplib.SMTP('smtp.mail.yahoo.com',465)
    elif emailid.partition("@")[2] == idlist[2]:
        server= smtplib.SMTP('smtp.live.com',587)
    else:
        print("only these mail id are accepted; gmail, outlook or yahoo")
    msg= mm('alternative')
    msg['From'] = emailid
    msg["To"] = To
    msg['subject'] = sub
    html = f"""
    {mesg}
    """
    text= bs4.BeautifulSoup(html,"html.parser").text
    text_part=txt(text,'plain')
    html_part=txt(html,'html')
    msg.attach(text_part)
    msg.attach(html_part)

    if (attachement):
        for file in attachement:
            f=open(file,'rb')
            data= f.read()
            attach_part = bse('application','octet-stream')
            attach_part.set_payload(data)
            email.encoders.encode_base64(attach_part)
            attach_part.add_header('Content-Disposition',f'attachment; filename= {file}')
            msg.attach(attach_part)
    print(f" This message: '{mesg}' has been sent to {To} ")
    #print(msg.as_string())
    #server= smtplib.SMTP('smtp.gmail.com',587)
    server.starttls()
    server.login(emailid,pswd)
    server.sendmail(From,To,msg.as_string())
    server.quit()
        
            
        
        
#===================================================================================================
emailid="pythonyser@gmail.com"
pswd= 'nomans10073'
From= "pythonyser@gmail.com"
To= "harisali.510@gmail.com"
sub="email using coding"
mesg="Hello how are you doing?"
attach=['test.log','text.txt','Untitled1.ipynb']
sendemail(emailid,pswd,To,sub,mesg,attach)
```

```python
#16. write a func to read a complete PDf file . 
from PyPDF4 import PdfFileReader
import os
def pdfreader(file_address):
    """
    
    pdfreader(file_address) read text inside of the pdf file
    file_address: (str) (location of the pdf file)
    eg: 'D:\\DESKTOP\\abc.pdf'
    
    """
    try:
        with open(file_address,'rb') as file:
            pdf=PdfFileReader(file)
            print('total no. of pages: ', pdf.getNumPages())
            for page in range(pdf.getNumPages()):
                print('PAGE NO:',page)
                ext=pdf.getPage(page)
                for i in ext.extractText():
                    print(i,end='')
    except Exception as e:
        print (e)
        
        
pdfreader('D:\\DESKTOP\\Scholarship application\\ALL DOC\\Motivation Letter\\MOTIVATION LETTER Aalborg.pdf')
```
